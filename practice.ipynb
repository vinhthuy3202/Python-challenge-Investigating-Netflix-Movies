```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix=pd.read_csv('datasets/netflix_data.csv')
netflix.head(5)
```

```python
netflix_movie=netflix[netflix['type']=='Movie']
netflix_movie_subset=netflix_movie[['title','country','release_year','duration','genre']]
netflix_movie_subset.head(5)
```

```python
fig=plt.figure(figsize=(12,8))
plt.scatter(netflix_movie['release_year'],netflix_movie['duration'])
plt.title('Movie Duration by Year of Release')
plt.xlabel('Release year')
plt.ylabel('Duration')
```

```python
short_movie=netflix_movie[netflix_movie['duration']<60]
short_movie_subset=short_movie[['title','country','genre','release_year','duration']]
short_movie_subset
```

```python
short_movie_subset['genre'].unique()
```

```python
color=[]
for i, row in short_movie_subset.iterrows():
    if row['genre']=='Children':
        color.append('Blue')
    elif row['genre']=='Documentaries':
        color.append('Green')
    elif row['genre']=='Stand-Up':
        color.append('Red')
    else:
        color.append('Yellow')
color[:10]
```

```python
plt.style.use('fivethirtyeight')
fig=plt.figure(figsize=(12,8))
plt.scatter(short_movie_subset['release_year'],short_movie_subset['duration'],c=color)
plt.title('Movie duration by year of release')
plt.xlabel('Release year')
plt.ylabel('Duration')
```
